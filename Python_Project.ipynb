```python
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.rename(columns={'Occupation': 'Sector'}, inplace=True)
```

```python
df['Marital_Status']=df['Marital_Status'].replace(0,'Single')
```

```python
df['Marital_Status']=df['Marital_Status'].replace(1, 'Married')
```

DV

```python
plt.figure(figsize=(8,5))
colorr = [ "#1F5C59", "#4B9C97", "#76B7B2", "#C2DEDC"]
c1 = sns.countplot(x='Gender', data=df, hue='Marital_Status', palette=colorr, edgecolor='black')

for a in c1.containers:
    c1.bar_label(a)

plt.xlabel('')
plt.ylabel('')

plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(x='Sector', y='Amount', data=df, errorbar=None, marker='o',color="#371C02")
plt.title('Sector Wise Amount',fontweight='bold',family='georgia',fontsize=16,color="#32281E")
plt.xlabel('')
plt.ylabel('')
plt.title('Amount Wise Sector')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
a = df.groupby('Zone', as_index=False)['Orders'].sum()
color_1 = [ "#E757E7", "#F178F1", "#EA0DA7", "#F288D7", "#EBC1E5"]
plt.pie(a['Orders'], labels=a['Zone'], autopct='%1.f%%', startangle=180, colors=color_1)
plt.title("Zone Wise Orders")
plt.show()
```

```python
plt.figure(figsize=(15,4))
color_2 = ["#2F4B6E", "#3A5C85", "#4E79A7", "#5F8DB8", "#6FA1C8","#81B2D6", "#93C3E2", "#A6D4ED", "#BBDFF4", "#D6ECFA"]


B = df.groupby('State',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
c3 = sns.barplot(x='State',y='Amount',data=B, palette=color_2)
plt.xticks(rotation=90)
#for s in c3.containers:
#    c3.bar_label(s) 
plt.show()


plt.figure(figsize=(15,4))
color_3 = [    "#2F5D2F", "#3E7A3E", "#4F9450", "#59A14F", "#6CB36A",
    "#80C686", "#9AD8A3", "#B6E7C1", "#D4F3DD", "#ECFAF0"
]
B = df.groupby('State',as_index=False)['Amount'].mean().sort_values(by='Amount',ascending=False).head(10)
c3 = sns.barplot(x='State',y='Amount',data=B, palette=color_3)
plt.xticks(rotation=90)
for s in c3.containers:
    c3.bar_label(s) 
plt.show()
```
